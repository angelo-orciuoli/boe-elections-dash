```python
pres = pd.read_csv('citywide_president_citywide.csv', header=None, usecols=[11, 12, 13, 14, 20, 21], names=['assembly_district', 'election_district', 'county', 'note', 'vote_choice', 'vote_count'])
pres['vote_count'] = pres['vote_count'].str.replace(',', '').astype(int)
pres['vote_choice'] = pres['vote_choice'].str.replace(r'\s*\(.*\)', '', regex=True)
pres = pres.groupby(['assembly_district', 'election_district', 'county', 'vote_choice'], as_index=False)['vote_count'].sum()
pres['vote_choice'] = pres['vote_choice'].replace({'Donald J. Trump / JD Vance': 'Trump', 'Kamala D. Harris / Tim Walz': 'Harris'})
pres
    
```
